```python
# 여러 데이터를 묶어서 그리기
import numpy as np

x = [1,2]
y1,y2,y3 = [1,2] , [2,5] , [3,7]

fig = plt.figure()
ax = fig.add_subplot(111)

w=0.2
ax.bar(x,y1,width=w , label='english')    #X가 1일때 빨강1, X가2일때 빨강2
ax.bar(np.array(x) + w, y2 , width=w , label = 'math')     #np.array(x) + w 는 [1 + w, 2 + w]이다.
ax.bar(np.array(x) + w*4, y3, width=w , label='Korean')

ax.legend()      #좌측상단애 나오는 범례 (설명)
plt.show()
```

```python
#바이올린 플롯에 평균/최소/최대 값 표시하기

import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)       # 정규분포 따르는 평균 0, 표편2.0, 표본 1000개 생성
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

#showmeans - 평균값 표시. (Default: False)
#showextrema - 최대/최소값 표시. (Default: True)
#showmedians는 중간값 표시. (Default: False)

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)

ax.set_ylim(-10.0, 10.0)    # y축 범위를 설정
ax.set_xticks(np.arange(1, 4))       # A,B,C 세 개를 나타내려고 X축에 눈금 1,2,3 표시
ax.set_xticklabels(['A', 'B', 'C'])  # 라벨링

ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# 바이올린 플롯 - 4분위수 표시하기
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], quantiles=[[0.25, 0.75], [0.1, 0.9], [0.3, 0.7]])

# quantiles 는 표시할 값을 지정함
# data_a - 25%, 75%의 위치에 표시, 
# data_b - 10%, 90%의 위치에 표시, 
# data_c - 30%, 70%의 위치에 표시

#그리고 data_c에 대해 30%, 70%의 위치에 분위수 직선을 표시합니다.
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
#바이올린 플롯 - 색상 지정하기
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('black')
violin['cmaxes'].set_edgecolor('red')
violin['cmins'].set_edgecolor('blue')
violin['cmeans'].set_edgecolor('yellow')

#violin[‘bodies’][n]의 set_facecolor() - n번째 바이올린 분포 영역의 색상 지정
#violin[‘cbars’]의 set_edgecolor() - 중심을 표시하는 직선의 색상 지정
#violin[‘cmaxes’]의 set_edgecolor() - 분포의 최대값을 표시하는 직선의 색상 지정
#violin[‘cmins’]의 set_edgecolor() - 분포의 최소값을 표시하는 직선의 색상 지정
#violin[‘cmeans’]의 set_edgecolor() - 분포의 평균을 표시하는 직선의 색상 지정

plt.show()
```
